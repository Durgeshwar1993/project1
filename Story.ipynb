# STORY #

STORY IS THAT REASON FOR  TRAFFIC VIOLATION OCCUR AND MAXIMUM NO. OF VIOLATION BY WHICH REASON AND PLOTTING GRAPH OF THIS STORY.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# LOAD DATA #

```python
df=pd.read_csv('/home/user/stack/Traffic_Violations.csv')
```

```python
descr1=pd.crosstab(df['Description'],df['Violation Type']).fillna(0)
descr1= descr1.sort_values(by=['Citation'],ascending=False , axis=0)
```

```python
descr1= descr1.iloc[0:15,0:4]
```

# GRAPH PLOT #

```python
descr1.plot.bar(stacked = False,figsize=(9,4))
plt.xlabel('Why violation occur',size =10)
plt.ylabel('No. of Violations',size=15)
plt.show()
```
