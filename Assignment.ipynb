# TRAFFIC VIOLATION #


```python
import pandas as pd
import matplotlib.pyplot as plt

```

# LOAD DATA #

```python
df=pd.read_csv('/home/user/stack/Traffic_Violations.csv')
```

# GRAPH PLOT #

```python

df['Violation Type'].value_counts().plot(kind='bar',color='m')
plt.ylabel('Count')
plt.xlabel('Violation Type')
plt.show()

```

# PART 2 #

# load data #

```python
df['Date Of Stop'] = pd.to_datetime(df['Date Of Stop'])
```

```python
df['Stop_Year'] = df['Date Of Stop'].dt.year
```

```python
locdf = pd.crosstab(df.Location,df.Stop_Year,).fillna(0)
```

```python
locdf1=locdf.sort_values(by=2013 , ascending=False)
```

```python
locdf2= locdf1.iloc[0:10,1:4]
```

# GRAPH PLOT #

```python


locdf2.plot.bar(figsize=(9,5))
plt.xlabel('Violation in years at diffrent places',size=10)
plt.ylabel('No. of Violations',size=15)
plt.show()
```

# PART 3 #

# Create DataFrame #

```python
group_subagency = pd.crosstab(df.SubAgency,df.Race)
```

# GRAPH PLOT #

```python

group_subagency.plot.bar(stacked=True,figsize=(10,10))
plt.xlabel('Violations first by District followed by Race', size =15)
plt.ylabel('No. of Violations', size=15)
plt.show()
```

# STORY #

```python
descr=pd.crosstab(df['Description'],df['Violation Type']).fillna(0)
descr1= descr.sort_values(by=['Citation'],ascending=False , axis=0)
```

```python
descr2= descr1.iloc[0:15,0:4]
```

```python
descr2.plot.bar(stacked = False,figsize=(9,4))
plt.xlabel('Why violation occur',size =10)
plt.ylabel('No. of Violations',size=15)
plt.show()
```
